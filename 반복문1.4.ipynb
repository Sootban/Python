```python
#Tutorial 1.3
'''
break

break는 흐름을 끊어 중단할 때 사용한다. 특정 코드를 반복하고자 
for, while 문을 이용하여 반복문(loop)을 만들었다면 break 문을 
사용하여 반복문을 중단할 수 있다.

반복문(loop) 순회 중에 break 문을 만나면 반복문의 내부 블록을 
벗어나게 된다. 그러나 예제 1번과 같이 조건문 안에 어떠한 장치 
없이 break를 넣으면 반복문이 반복하지 않으므로 예제 2번과 같이 
주로 조건 안에 넣어 실행한다.

예제 1)
<1>
while 조건:
    반복 실행할 코드
    break # while 구문을 탈출함

<2>    
for 변수 in 범위:
    break # for 반복문 탈출
    
예제 2)
<1>
while 조건:
    반복 실행할 코드
    if 조건:
        break # while 구문을 탈출
        
<2>
for 변수 in 범위:
    반복 실행할 코드
    if 조건:
        break # for 반복문 탈출
'''
```

```python
#Tutorial 1.4
'''
continue

continue의 사전적 의미를 살펴보면 '계속하다', '계속가다'라는 뜻이 
있다. 파이썬에서 continue문은 반복문이 실행하는 코드블록의 나머지 
부분을 실행하지 않고 다음 반복으로 건너가게 흐름을 조정한다.

#continue와 pass는 구분해서 사용을 해야 하니 차이점을 잘 정리해두자.

반복문 순회 도중 continue문을 만날시 continue문은 이후의 나머지 반복문 
내부 코드블록을 실행하지 않고 다음 아이템을 선택해 반복문의 내부 코드블
록 시작부분으로 이동한다. break와 마찬가지로 조건이 없이 continue를 사
용하는 것은 큰 의미가 없으므로(예제 1) 주로 조건문을 달아 실행(예제2) 
한다.

예제 1)
<1>
while 조건:
    반복 실행할 코드
    continue # while 다음 반복문 수행
    continue 사용시 무시되는 코드

<2>
for 변수 in 범위:
    반복 실행할 코드
    continue # for 다음 반복문 수행
    continue 사용시 무시되는 코드

예제 2)
<1>
while 조건:
    반복 실행할 코드
    if 조건:
        continue # while 다음 반복문 수행
    continue 사용시 무시되는 코드

<2>
for 변수 in 범위:
    반복 실행할 코드
    if 조건:
        continue # for 다음 반복문 수행
    continue 사용시 무시되는 코드
'''
```

```python
'''
<code>
수험번호_점수 = [
    (1001, 90),
    (1002, 80),
    (1003, 70),
    (1004, 60),
    (1005, 58),
    (1006, 90),
    ]
for 수험번호, 점수 in 수험번호_점수:
    if 점수 < 60
        continue
    print('수험번호 {}님 {}점으로 합격하셨습니다.'.format(수험번호, 점수))

점수가 60점 미만인 학생의 경우에는 점수 < 60이 참(True)가 되어 continue문이 
실행된다. 따라서 나머지 코드블록인 합격 메시지를 출력해주는 print문을 실행하
지 않고 다시 for 반복문의 내부 코드블록 시작부분으로 돌아간다.

여기서 continue라고 되어 있는 부분을 pass로 바꾸면 수험번호 1005번의 58점도 
출력하는것을 볼 수 있다. 반복 중간에 continue를 만나면 다음 반복으로 넘어가
지만 pass를 만나면 그 라인만 지나치고 아래 문장을 그대로 실행한다.
'''
```

```python
#Tutorial 1.5
'''
else

앞서 살펴본 것처럼 Python에서는 while, for 문에서도 else 문을 사용할 
수 있다. 여기서의 else는 if에서의 else처럼 '그렇지 않으면'이라는 의미 
보다는 '그런 다음'이라는 의미가 더 강하기 때문에 than으로 스여야 된다
는 논의가 있기도 했다.

반복문이 break등에 의해 중단없이 정상적으로 반복이 종료된후 특정 코드를 
실행하게 해야할 때 while~ else, for~ else를 사용할 수 있다. 또한 if문에서 
else는 '그렇지 않으면'이라는 의미로 조건이 거짓일 때 쓰이고, 
try~except~else문에서 else는 예외가 발생하지 않을 때 쓰인다.
'''
```

```python
#Tutorial 1.6
'''
중첩반복문

반복문 내부에 또 다른 반복문이 있을 경우, 중첩되었다고 얘기한다. 
중첩반복문이란 반복문 안에 반복문이 들어가 중첩된 것을 얘기한다. 
중첩 반복문에서는 종료되는 값을 항상 확인하자.

<code>
i = 2
while i < 10:
    k = 1
    while k < 10:
        print(i, "*", k, "=", i * k)
        k += 1
    i += 1

구구단은 2단부터 시작하기에 변수 i를 2로 설정해준다. 9단을 수행한 
후 종료할 수 있도록 10 미만이라는 조건을 설정해주었다. i는 구구단의 
단을, k는 순차적으로 반복될 곱의 수이며 while문을 하나 더 이용하여 
반복될 곱의 수를 1씩 순차적으로 더해주었다.

여기서 중요한 것은 종료되는 값이다. k가 10이 된 상태로 안에 반복문이 
종료되기 때문에 3단을 하기 위해서는 다시 k를 1로 초기화 시켜줘야 한다.

for문과 range 함수를 이용하면 좀 더 간단하게 구구단을 출력할 수 있다.
'''
```

```python
i = 2
while i < 10:
    k = 1
    while k < 10:
        print(i, "*", k, "=", i * k)
        k += 1
    i += 1
```

```python
'''
<code>
for i in range(2, 10):
    print("---{0}단---".format(i))
    for k in range(1, 10):
        print(i, "*", k, "=", i * k)

for문을 중첩 사용하면 좀 더 간결하고 직관적인 코드를 작성할 수 있다.

첫번째 for문에서 range함수를 이용하여 2단부터 9단까지 차례로 i에 대입된다.
두번째 for문에서 곱해지는 수를 k에 넣고 구구단을 출력하고 있다.

while과 달리 k의 값이 for문 안에서 자동으로 초기화가 되기 때문에 
while문처럼 k = 1 을 두번째 for문 위에 넣을 필요는 없다.
'''
```
